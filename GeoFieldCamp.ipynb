```python
import networkx as nx
import matplotlib.pyplot as plt

Strati=["Sebastian\nGarzón","Daniel\nGongora"]  
PaleoMag=["Esteban\nGaitan","Alejandro\nRodriguez"]
Geofisica=["Billy\nRevelo","Andrea\nJara","Juliana\nAyala"]
Petrologia=["Nathalia\nPineda","Nicolas\nZuluaga","Paz\nUrdaneta","Carolina\nRubiano","Margarita\nAriza",
           "Catalina\nAgudelo"]
GeoStructural=["Carolina\nOrtiz","Jose\nAvellaneda","Lina\nPerez","Juan\nMeza"]


G=nx.Graph()
G.add_nodes_from(Strati)
G.add_nodes_from(PaleoMag)
G.add_nodes_from(Geofisica)
G.add_nodes_from(Petrologia)
G.add_nodes_from(GeoStructural)

for name in Strati:
    G.add_edge("GeoField\nCamp",name)
for name in PaleoMag:
    G.add_edge("GeoField\nCamp",name)
for name in Geofisica:
    G.add_edge("GeoField\nCamp",name)
for name in Petrologia:
    G.add_edge("GeoField\nCamp",name)
for name in GeoStructural:
    G.add_edge("GeoField\nCamp",name)
    
plt.figure(figsize=(7,7))
plt.title("$Trabajos\ de\ grado$",size=25)
StratiDic = {name: 0.2 for name in Strati}
PaleoDic={name:0.3 for name in PaleoMag}
GeofiDic={name:0.5 for name in Geofisica}
PetroDic={name:0.6 for name in Petrologia}
StrucDic={name:0.70 for name in GeoStructural}
d4 = dict(StratiDic.items()| PaleoDic.items()|GeofiDic.items()|PetroDic.items()|StrucDic.items())
d4["GeoField\nCamp"]=0.1
values = [d4.get(node, 1) for node in G.nodes()]

nx.draw(G, cmap=plt.cm.get_cmap('RdBu'), node_color=values,with_labels=True,node_size=3200,font_size=9,font_color='white')
```
